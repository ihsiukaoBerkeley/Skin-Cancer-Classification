### Adding image paths to metadata df

```python
folder1_path = '/Users/nathanarias/Desktop/MIDS/207/final_project/archive/HAM10000_images_part_1'
folder2_path = '/Users/nathanarias/Desktop/MIDS/207/final_project/archive/HAM10000_images_part_2'

# Function to get the full path of an image file, trys both folders in the archive directory
def get_image_path(image_id, folder1_path, folder2_path):
    file_name = image_id + '.jpg'
    
    path1 = os.path.join(folder1_path, file_name)
    path2 = os.path.join(folder2_path, file_name)

    if os.path.exists(path1):
        return path1
    elif os.path.exists(path2):
        return path2
    else:
        return None  # Or handle this as you see fit

# Adding the image path to the dataframe
meta_data['image_path'] = meta_data['image_id'].apply(lambda x: get_image_path(x, folder1_path, folder2_path))


#checking success
meta_data.image_path.isna().sum()
```

# Looking at different `dx` classes

```python
#creating label names list
label_names = list(train_eda.dx.unique())
label_names
```

```python
#a helper function for finding the first n indices of a particular label
def index_finder(labels_list, target, n):
    """Find first n indices in labels_list that match target."""
    label_indices = []
    
    for index, label in enumerate(labels_list):
        if label == target:
            label_indices.append(index)
            if len(label_indices) == n:
                break
                
    return label_indices


#displaying the first 5 entrys of each dx type
fig, axs = plt.subplots(nrows=7, ncols=5, figsize=(12,12))

for i in range(7):
    image_indices = index_finder(y_train, target=label_names[i], n=5)
    for j in range(5):
        img_path = train_eda.loc[image_indices[j], 'image_path']
        image = Image.open(img_path) 
        label_name = label_names[i]
        
        axs[i][j].imshow(image)
        axs[i][j].set_title(label_name)
        axs[i][j].axis('off')

plt.show()
```

```python
#creating a label dictionary and applying to the y_train labels
label_dict = {'nv':0,'mel':1,'vasc':2,'bcc':3,'bkl':4,'akiec':5,'df':6}
y_train_indices = np.vectorize(label_dict.get)(y_train)
```

```python
X_train_image_paths = []
for i in X_train:
    X_train_image_paths.append(i[6])
    
```

```python
# Split proportions
split = [0.9, 0.1]

# Calculate the number of elements in each split
splits = np.multiply(len(X_train_image_paths), split).astype(int)

# Correct way to use np.split
train_x, val_x = np.split(X_train_image_paths, [splits[0]])

len(train_x), len(val_x)
```

```python
def load_and_preprocess(path):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [img_height, img_width])
    image /= 255.0
    return image

# Define image dimensions
img_width, img_height = 224, 224

# making a tf_dataset
X_train_dataset = tf.data.Dataset.from_tensor_slices(train_x)
#X_val_dataset = tf.data.Dataset.from_tensor_slices(val_x)
Y_train_dataset = tf.data.Dataset.from_tensor_slices(y_train_indices)

# Mapping load-and process
X_train_dataset = X_train_dataset.map(load_and_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
#X_val_dataset = X_val_dataset.map(load_and_preprocess, num_parallel_calls= tf.data.AUTOTUNE)

X_train_dataset, Y_train_dataset
```
